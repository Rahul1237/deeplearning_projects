```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
f = pd.read_csv("C:\\Users\\dell\\Downloads\\Compressed\\Churn_Modelling.csv")
df = pd.DataFrame(f)
```

```python
df
```

```python
dummies = pd.get_dummies(df.Geography)
```

```python
df_1 = pd.concat([df,dummies],axis='columns')
```

```python
df_1
```

```python
df = df_1.drop(['Geography'],axis = 'columns')
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
labelencoder = LabelEncoder()
```

```python
df.Gender = labelencoder.fit_transform(df.Gender)
```

```python
df
```

```python
y = df.Exited
```

```python
X = df.drop(['RowNumber','CustomerId','Surname','Exited'],axis = 'columns')
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2,random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
X_train = sc.fit_transform(X_train)
```

```python
X_test = sc.fit_transform(X_test)
```

```python
import tensorflow as tf
```

```python
classifier = tf.keras.models.Sequential()
```

```python
classifier.add(tf.keras.layers.Dense(units=6,activation='relu'))
```

```python
classifier.add(tf.keras.layers.Dense(units=6,activation='relu'))
```

```python
classifier.add(tf.keras.layers.Dense(units=1,activation='sigmoid'))
```

```python
classifier.compile(optimizer='adam',loss='binary_crossentropy',metrics = ['accuracy'])
```

```python
classifier.fit(X_train,y_train,batch_size=30,epochs = 100)
```

```python
y_pred = classifier.predict(X_test)
```

```python
y_pred = (y_pred > 0.5)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm = confusion_matrix(y_test,y_pred)
```
